    It decides whether the values are equal or not.If both the values are equal then it returns true. for example:-

```python
a=10
b=10
a==b
```

```python
a=5
b=4
a>b
```

```python
a=5
b=6
a<b
```

```python
a = 10
b = 10 
a >= b 
```
